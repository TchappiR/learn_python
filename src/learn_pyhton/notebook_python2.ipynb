```python
class Vehicule:
    
    def __init__(self, marque : str, vitesse_max : float):
        self._marque = marque
        self.__vitesse_max = vitesse_max
    
    def __str__(self) -> str:
        return f"Vehicule de marque : {self._marque}"
    
    def __eq__(self, other_vehicule : "Vehicule") -> bool:
        return self._marque.lower() == other_vehicule._marque.lower() and self.__vitesse_max == other_vehicule.__vitesse_max

    def deplacer(self):
        return f"Le vehicule {self._marque} se déplace ..."
```

```python
class Voiture(Vehicule):
    def __init__(self, marque : str, vitesse_max : float, portes : int):
        super().__init__(marque, vitesse_max)
        self.portes = portes

    def deplacer(self):
        return f"Le voiture {self._marque} se route à {self.portes} portes ..."

class Moto(Vehicule):
    def deplacer(self):
        return f"La moto {self._marque} fonce ..."
```

```python
vehicules = [
    Voiture("Peugeot", 180.0, 4),
    Moto("Honda", 200.0)
]
```

```python
for v in vehicules:
    print(v.deplacer())
```

```python
v1 = Voiture("Toyota", 150.0, 4)
v2 = Voiture("Toyota", 150.0, 5)
```

```python
print(v1 == v2)  
```
